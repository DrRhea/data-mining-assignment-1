#### Visualisasi Data

```python
# Data
categories = sekolah_merged['Kabupaten/Kota'].head(6).tolist()
sd_negeri = sekolah_merged['Jumlah Murid SD (Negeri)'].head(6).tolist()
sd_swasta = sekolah_merged['Jumlah Murid SD (Swasta)'].head(6).tolist()
smp_negeri = sekolah_merged['Jumlah Murid SMP (Negeri)'].head(6).tolist()
smp_swasta = sekolah_merged['Jumlah Murid SMP (Swasta)'].head(6).tolist()
sma_negeri = sekolah_merged['Jumlah Murid SMA (Negeri)'].head(6).tolist()
sma_swasta = sekolah_merged['Jumlah Murid SMA (Swasta)'].head(6).tolist()
smk_negeri = sekolah_merged['Jumlah Sekolah SMK (Negeri)'].head(6).tolist()
smk_swasta = sekolah_merged['Jumlah Sekolah SMK (Swasta)'].head(6).tolist()


# Bar chart settings
x = np.arange(len(categories))
width = 0.1


# Plot
fig, ax = plt.subplots(figsize=(12, 8))


# Add bars
ax.bar(x - 3*width, sd_negeri, width, label='SD Negeri', color='skyblue')
ax.bar(x - 2*width, sd_swasta, width, label='SD Swasta', color='blue')
ax.bar(x - width, smp_negeri, width, label='SMP Negeri', color='lightgreen')
ax.bar(x, smp_swasta, width, label='SMP Swasta', color='green')
ax.bar(x + width, sma_negeri, width, label='SMA Negeri', color='orange')
ax.bar(x + 2*width, sma_swasta, width, label='SMA Swasta', color='red')
ax.bar(x + 3*width, smk_negeri, width, label='SMK Negeri', color='purple')
ax.bar(x + 4*width, smk_swasta, width, label='SMK Swasta', color='brown')


# Labels and title
ax.set_xlabel('Kabupaten/Kota')
ax.set_ylabel('Jumlah Murid')
ax.set_title('Jumlah Murid SD, SMP, SMA, SMK per Kabupaten/Kota di DKI Jakarta pada Tahun 2022')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()


plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

- Kepulauan Seribu memiliki jumlah murid yang paling sedikit karena populasinya yang kecil, sementara Kota Jakarta Timur dan Jakarta Barat memiliki jumlah murid yang paling tinggi

```python
# Membaca data
data_daerah = pd.read_csv("https://drive.google.com/uc?export=download&id=1PughxjXW8_Sh8l16NBY57of6CzFv51Vs")

# Membuat figure untuk heatmap saja
fig, ax1 = plt.subplots(figsize=(12, 8))

# Normalisasi data untuk setiap daerah
normalized_data = data_daerah.copy()
regions = ['Kep Seribu', 'Jakarta Selatan', 'Jakarta Timur',
           'Jakarta Pusat', 'Jakarta Barat', 'Jakarta Utara']

for region in regions:
    total = data_daerah[region].sum()
    normalized_data[region] = (data_daerah[region] / total) * 100

# Membuat heatmap
sns.heatmap(normalized_data[regions].T,
            annot=True,
            fmt='.1f',
            cmap='YlOrRd',
            ax=ax1,
            cbar_kws={'label': 'Persentase Penduduk (%)'},
            xticklabels=normalized_data['Umur'])

# Menyesuaikan tampilan heatmap
ax1.set_xticklabels(normalized_data['Umur'], rotation=45)
ax1.set_title('Distribusi Umur per Daerah (Normalized %)', pad=20)
ax1.set_xlabel('Kelompok Umur')
ax1.set_ylabel('Daerah')

# Menampilkan plot
plt.show()
```

Dapat kita liat pada heat map diatas, daerah-daerah di Jakarta memiliki persentase yang relatif tinggi pada kelompok usia 0-14 tahun, dengan angka yang menurun seiring bertambahnya usia.

Variasi Antar Wilayah:
*   Kepulauan Seribu memiliki persentase tertinggi di kelompok usia 0-4 tahun, sedangkan daerah lain cenderung memiliki persentase lebih merata di rentang usia muda dan dewasa.
*   Jakarta Barat dan Jakarta Utara memiliki pola distribusi yang sedikit lebih merata dibandingkan dengan Jakarta Selatan dan Kepulauan Seribu.
*   Jakarta Barat dan Jakarta Utara memiliki pola distribusi yang sedikit lebih merata dibandingkan dengan Jakarta Selatan dan Kepulauan Seribu.
*   Kelompok Usia Tua (60+ tahun) memiliki persentase yang relatif rendah di semua wilayah, menunjukkan bahwa populasi lansia lebih kecil dibandingkan dengan kelompok usia lainnya.

```python
# Set figure size
plt.figure(figsize=(12, 6))


# Set bar width and position
bar_width = 0.35
index = np.arange(len(data_jk['Jenis kelamin'][1:]))


# Plot bar chart for 2022 Laki-laki and Perempuan
plt.bar(index, data_jk['Laki-laki.1'][1:], bar_width, label='Laki-laki', color='blue')
plt.bar(index + bar_width, data_jk['Perempuan.1'][1:], bar_width, label='Perempuan', color='pink')


# Add labels and title
plt.xlabel('Kelompok Umur')
plt.ylabel('Jumlah Penduduk')
plt.title('Jumlah Penduduk Berdasarkan Kelompok Umur dan Jenis Kelamin di DKI Jakarta (2022)')
plt.xticks(index + bar_width / 2, data_jk['Jenis kelamin'][1:], rotation=45)


# Add legend
plt.legend()


# Show the plot
plt.tight_layout()
plt.show()
```

### Tren Daya Beli Masyarakat
#### Load Dataset
Kami melakukan load dataset dari dua file berbeda, yaitu file dataset Pengeluaran per Kapita pada tahun 2019 - 2021 dan file dataset Pengeluaran per Kapita pada tahun 2022.
#### sumber : data.go.id dan diakses pada 01/11/2024


```python
# Load dataset
file_path = 'https://drive.google.com/uc?export=download&id=1zgMECuU2ypf_7Kle57G06KX2ePZQ7kTB'
data_1 = pd.read_csv(file_path)


file_path_2 = 'https://drive.google.com/uc?export=download&id=1wBIkG2RiTWvT_A4N9ht7aYBRB0ICEBOo'
data_2 = pd.read_csv(file_path_2)
```

#### Merge Dataset
Setelah itu, kami melakukan merge dataset berdasarkan Kab/kota, sehingga dataset tersebut berisi Pengeluaran per Kapita mulai dari tahun 2019 - 2022.

```python
# Merge dataset berdasarkan kab/kota
gabungan_data = pd.merge(data_1, data_2, on="Kab/Kota")

gabungan_data.head()
```

#### Cleaning Dataset
Setelah itu, kami membersikan dataset, jika data memiliki header tambahan atau baris kosong. Kami akan menghapus baris yang tidak diperlukan atau tidak relevan serta mengubah nama kolom menjadi lebih jelas.

```python
# Membersihkan data: misalnya, jika data memiliki header tambahan atau baris kosong
# Menghapus baris pertama yang tidak diperlukan, dan memberikan nama kolom yang jelas
data_cleaned = gabungan_data.dropna()  # Menghapus baris kosong
data_cleaned.columns = ["Kab/Kota", "2019", "2020", "2021", "2022"]  # Mengganti nama kolom

# Menampilkan beberapa baris untuk memastikan data telah dibersihkan dengan benar
data_cleaned.head()
```

#### Memvisualisasikan Dataset menggunakan Diagram
Pertama-tama, kami memilih beberapa wilayah yang akan dimasukkan kedalam diagram dan akan memisahkan data yang akan kita masukkan dengan yang tidak.

```python
# Memilih beberapa wilayah untuk divisualisasikan
wilayah_terpilih = ["Kep Seribu", "Jakarta Utara", "Jakarta Timur", "Jakarta Pusat", "Jakarta Selatan", "Jakarta Barat"]


# Menyaring data untuk wilayah yang terpilih
data_filtered = data_cleaned[data_cleaned["Kab/Kota"].isin(wilayah_terpilih)]
```

Setelah itu, kami melakukan pengubahan tipe data kedalam bentuk numerik agar dapat divisualisasikan.

```python
# Assuming data_filtered is your filtered DataFrame
data_filtered.loc[:, "2019"] = pd.to_numeric(data_filtered["2019"], errors='coerce')
data_filtered.loc[:, "2020"] = pd.to_numeric(data_filtered["2020"], errors='coerce')
data_filtered.loc[:, "2021"] = pd.to_numeric(data_filtered["2021"], errors='coerce')
data_filtered.loc[:, "2022"] = pd.to_numeric(data_filtered["2022"], errors='coerce')
```

Setelah itu, kita akan membuat diagram berdasarkan data yang telah kita proses tadi.

```python
plt.figure(figsize=(10,6))


for index, row in data_filtered.iterrows():
    plt.plot(["2019", "2020", "2021", "2022"], row[["2019", "2020", "2021", "2022"]], marker='o', label=row["Kab/Kota"])


# Menambahkan judul dan label
plt.title("Tren Pengeluaran Per Kapita (2019 - 2022) untuk Wilayah Terpilih", fontsize=14)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Pengeluaran Per Kapita", fontsize=12)
plt.legend(title="Wilayah")
plt.grid(True)


# Menampilkan plot
plt.show()
```

#### Kesimpulan
Berdasarkan hasil visualisasi data mengenai pengeluaran per kapita di provinsi DKI Jakarta dari tahun 2019 hingga 2022, dapat disimpulkan bahwa terdapat ketidakmerataan dalam distribusi pengeluaran antar wilayah. Salah satu contoh yang mencolok terlihat pada data tahun 2022, di mana Kepulauan Seribu mencatatkan pengeluaran per kapita yang sangat rendah, yakni sekitar Rp 12.999,00, sementara Jakarta Selatan, yang merupakan salah satu wilayah dengan pengeluaran tertinggi, mencapai sekitar Rp 24.221,00.

Ketidakmerataan pengeluaran per kapita ini dapat berdampak serius pada kualitas hidup masyarakat. Wilayah dengan pengeluaran rendah seperti Kepulauan Seribu berpotensi menghadapi masalah seperti kemiskinan, kurangnya akses pendidikan, dan layanan kesehatan yang terbatas. Sementara itu, wilayah dengan pengeluaran tinggi seperti Jakarta Selatan mungkin mengalami tingkat kesejahteraan yang lebih baik, tetapi juga dapat menimbulkan masalah sosial seperti segregasi ekonomi dan kurangnya solidaritas sosial.

### Layanan Kesehatan

```python
# Create DataFrame
data = {
    'Daerah': ['Kep Seribu', 'Kep Seribu', 'Kep Seribu', 'Jakarta Selatan', 'Jakarta Selatan', 'Jakarta Selatan',
               'Jakarta Timur', 'Jakarta Timur', 'Jakarta Timur', 'Jakarta Pusat', 'Jakarta Pusat', 'Jakarta Pusat',
               'Jakarta Barat', 'Jakarta Barat', 'Jakarta Barat', 'Jakarta Utara', 'Jakarta Utara', 'Jakarta Utara'],
    'Tahun': [2021, 2022, 2023, 2021, 2022, 2023, 2021, 2022, 2023, 2021, 2022, 2023, 2021, 2022, 2023, 2021, 2022, 2023],
    'Rumah Sakit': [1, 1, 1, 53, 57, 57, 46, 54, 54, 36, 38, 38, 27, 29, 29, 27, 30, 30],
    'Tempat Tidur': [14, 10, 10, 6720, 4745, 4745, 6362, 6104, 6104, 6349, 5662, 5662, 5111, 4055, 4055, 3999, 3091, 3091],
    'Puskesmas Kecamatan': [2, 2, 2, 10, 10, 10, 10, 10, 10, 8, 8, 8, 8, 8, 8, 6, 6, 6],
    'Puskesmas Kelurahan': [6, 4, 4, 69, 67, 67, 77, 75, 75, 36, 34, 34, 65, 67, 67, 43, 41, 41]
}

df = pd.DataFrame(data)

# Calculate mean per region
mean_per_region = df.groupby('Daerah').mean(numeric_only=True).drop(columns='Tahun')

# Create normalized version of the data
normalized_data = mean_per_region.copy()

# Normalize Tempat Tidur by dividing by 100 to make it more comparable
normalized_data['Tempat Tidur'] = normalized_data['Tempat Tidur'] / 100


# Buat subplot dengan dua sumbu, tetapi hanya gunakan ax2
fig, ax2 = plt.subplots(1, 1, figsize=(12, 6))  # Sesuaikan ukuran sesuai kebutuhan

# Original plot
bar_width = 0.2
x = np.arange(len(mean_per_region.index))

# Normalized plot
ax2.bar(x - bar_width * 1.5, normalized_data['Rumah Sakit'], bar_width, label='Rumah Sakit')
ax2.bar(x - bar_width/2, normalized_data['Tempat Tidur'], bar_width, label='Tempat Tidur (100)')
ax2.bar(x + bar_width/2, normalized_data['Puskesmas Kecamatan'], bar_width, label='Puskesmas Kecamatan')
ax2.bar(x + bar_width * 1.5, normalized_data['Puskesmas Kelurahan'], bar_width, label='Puskesmas Kelurahan')

ax2.set_xlabel('Region')
ax2.set_ylabel('Average (Normalized Scale)')
ax2.set_title('Healthcare Facilities by Region (Normalized Scale)')
ax2.set_xticks(x)
ax2.set_xticklabels(normalized_data.index, rotation=45)
ax2.legend()

plt.tight_layout()
plt.show()
```

Berdasarkan diagram diatas, Setiap wilayah memiliki jumlah puskesmas kelurahan (merah) yang dominan, terutama di Jakarta Timur yang memiliki nilai tertinggi. Tempat tidur di rumah sakit (oranye) juga relatif banyak di seluruh wilayah, dengan jumlah terbanyak di Jakarta Selatan. Rumah sakit (biru) tersebar di seluruh wilayah dengan jumlah tertinggi di Jakarta Selatan dan Jakarta Timur. Puskesmas kecamatan (hijau) jumlahnya paling sedikit dibandingkan fasilitas lainnya dan hampir merata di semua wilayah, kecuali Kepulauan Seribu yang memiliki fasilitas kesehatan yang jauh lebih sedikit dibandingkan wilayah lain.
